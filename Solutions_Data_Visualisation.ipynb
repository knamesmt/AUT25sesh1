### Data Visualisation

“The simple graph has brought more information to the data analyst’s mind than any other device.” — John Tukey

We will use the London Bikes dataset, showing the number of bikes rented every day from 2010-2024, including variables such as temp, humidity etc

**Import the necessary libraries and dataset now**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

bikes = pd.read_csv('../Data/london_bikes.csv')
bikes

# Convert to datetime first
bikes['date'] = pd.to_datetime(bikes['date'])

# extract components
bikes["year"] = bikes["date"].dt.year
bikes["month"] = bikes["date"].dt.month
bikes["month_name"] = bikes["date"].dt.month_name()   # month label
bikes["day"] = bikes["date"].dt.day
bikes["day_of_week"] = bikes["date"].dt.day_name()     # weekday label
bikes['date'] = bikes['date'].dt.date


# create season_name based on numeric rules
bikes["season"] = np.select(
    [
        (bikes["month"].isin([12, 1, 2])),   # Winter: Dec, Jan, Feb
        (bikes["month"].isin([3, 4, 5])),    # Spring: Mar, Apr, May
        (bikes["month"].isin([6, 7, 8])),    # Summer: Jun, Jul, Aug
        (bikes["month"].isin([9, 10, 11]))   # Autumn: Sep, Oct, Nov
    ],
    ["Winter", "Spring", "Summer", "Autumn"], 
    default="Unknown"
)

# convert to ordered categorical, like fct_relevel in R
season_order = ["Winter", "Spring", "Summer", "Autumn"]
bikes["season"] = pd.Categorical(
    bikes["season"],
    categories=season_order,
    ordered=True
)

bikes

bikes
```

#### Histogram plot

```python
# Create histogram of bikes hired
plt.figure(figsize=(8,5))
sns.histplot(bikes["bikes_hired"], bins=30, kde=False, color="steelblue")

# add titles and tidy style
plt.title("Distribution of Bikes Hired")
plt.xlabel("Bikes Hired")
plt.ylabel("Frequency")
plt.grid(True, linestyle="--", alpha=0.5)  # similar to theme_bw()
plt.show()
```

```python
# create faceted histograms by month_name
g = sns.FacetGrid(bikes, col="month", col_wrap=4, sharex=False, sharey=False)
g.map_dataframe(sns.histplot, x="bikes_hired", bins=30)

# add titles and tidy style
g.set_axis_labels("Bikes Hired", "Frequency")
g.set_titles(col_template="{col_name}")  # each facet labelled by month
plt.tight_layout()
plt.show()
```

**Density Plot**

```python
# Density plots of bikes hired by month and season
g = sns.FacetGrid(bikes, col="season", col_wrap=4, sharex=True, sharey=True)
g.map_dataframe(sns.kdeplot, x="bikes_hired", fill=True, alpha=0.3, linewidth=0)

# add titles and tidy style
g.set_axis_labels("Bikes Hired", "Density")
g.set_titles("{col_name}")
plt.tight_layout()
plt.show()
```

**Boxplots**

```python
# Boxplots of bikes hired by month and season
plt.figure(figsize=(12,6))
sns.boxplot(data=bikes, x="month_name", y="bikes_hired", hue="season")

# add titles and tidy style
plt.xlabel("Month"); plt.ylabel("Bikes Hired")
plt.legend(title="Season")
plt.tight_layout()
plt.show()
```

```python
# Scatter plots of bikes hired vs mean_temp, faceted by season
g = sns.lmplot(
    data=bikes,
    x="mean_temp", y="bikes_hired",
    col="season", col_wrap=4,
    scatter_kws=dict(alpha=0.2, s=15),
    line_kws=dict(color="black"),
    ci=None  # se = FALSE
)

# add titles and tidy style
g.set_axis_labels("Average Temperature (°C)", "Bikes Hired")
plt.tight_layout()
plt.show()
```

**Time Series**

```python
# 30-day rolling mean as the smooth 
bikes_sorted = bikes.sort_values("date").copy() # ensure sorted by date
bikes_sorted["trend_30d"] = bikes_sorted["bikes_hired"].rolling(30, min_periods=1).mean() 

# Time series plot with trend line
plt.figure(figsize=(12,5))
plt.scatter(bikes_sorted["date"], bikes_sorted["bikes_hired"], alpha=0.5, s=8)
plt.plot(bikes_sorted["date"], bikes_sorted["trend_30d"], color="red", linewidth=1.5)
plt.xlabel("Date"); plt.ylabel("Bikes Hired")
plt.tight_layout(); plt.show()
```
